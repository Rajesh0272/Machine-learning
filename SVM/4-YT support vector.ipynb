```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.data
```

```python
iris.target_names
```

```python
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df
```

```python
iris_df['target']=iris.target
iris_df.head()
```

```python
iris_df['flower_names']=iris_df.target.apply(lambda x: iris.target_names[x])
iris_df
```

```python
df0=iris_df[:50]
df1=iris_df[50:100]
df2=iris_df[100:150]
```

```python
from sklearn.model_selection import train_test_split
x=iris_df.drop(['target','flower_names'],axis='columns')
y=iris_df.target

x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.datasets import load_digits
ld=load_digits()
```

```python
ld.target
```

```python
digits=pd.DataFrame(ld.data)
digits
```

```python
digits['target']=ld.target
```

```python
digits.head(15)
```

```python
from sklearn.model_selection import train_test_split
x=digits.drop('target',axis='columns')
y=digits.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
sep=SVC(kernel='rbf')
```

```python
sep.fit(x_train,y_train)
```

```python
y_pred=sep.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
